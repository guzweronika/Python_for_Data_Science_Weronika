```python
"""Creating catefgorical variables and get IDs with null value."""
import pandas as pd

car_colors = pd.Series(["Blue", "Red", "Green"], dtype = 'category')

car_data = pd.Series(pd.Categorical(["Yellow", "Green", "Red", "Blue", "Purple"], categories=car_colors, ordered = False))

find_entries = pd.isnull(car_data)

print("Car colors")
print(car_colors)
print()
print("Car data")
print(car_data)
print()
print("NUll colors")
print(find_entries[find_entries == True])
```

```python
"""Rename the categories. Renaming the data is doing with the new pattern."""
"""When I add vdata which is not in the first pattern I get NaN - The size of letter. spaces and deviation from pattern do matter."""

import pandas as pd

car_colors = pd.Series(["Blue", "Red", "Green"], dtype = 'category')

car_data = pd.Series(pd.Categorical(["Blue", "Green", "Red", "Blue", "Red", "Red", "Red", "Green", "Yellow", " Red", "red", "Geld"], categories = car_colors, ordered = False))

print("Car colors before rename the categories.")
print(car_colors)
print("\nCar data before rename the categories.")
print(car_data)

# cat.categories - set a new value
car_colors.cat.categories = ["Purple", "Yellow", "Mauve"]
car_data.cat.categories = car_colors                                    

print("\n\nCar data after rename the categories.")
print(car_data)
print("\nCar colors after rename the categories.")
print(car_colors)
```

```python
import pandas as pd

car_colors = pd.Series(["Blue", "Red", "Green"], dtype='category')
car_data = pd.Series(pd.Categorical(["Blue", "Green", "Red", "Green", "Red", "Green"], categories=car_colors, ordered=False))

car_data.cat.categories = ["Blue_Red", "Red", "Green"]

print("Print car data.")
print(car_data)

red = car_data.isin(["Red"])

print("\nPrint car data where color is Red")
print(car_data.loc[red])

car_data.loc[red] = "Blue_Red"
print("\nPrint car data after changing Red to Blue_Red")
print(car_data)
```
