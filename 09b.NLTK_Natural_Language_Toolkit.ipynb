```python
# https://towardsai.net/p/nlp/natural-language-processing-nlp-with-python-tutorial-for-beginners-1f54e610a1a0

# Open the test file
text_file = open("NLP_text.txt")
```

```python
# Read the data
text = text_file.read()

# Data type
print(f"\nData type: {type(text)}")

# Text
print(f"\nTEXT:\n\n{text}")

# Length of the text
print(f"\nLength of the text: {len(text)}")
```

```python
import nltk
from nltk import sent_tokenize, word_tokenize

# Tokenize the text by sentences
sentences = sent_tokenize(text)
print(f"Number of sentences: {len(sentences)}")
sentences
```

```python
# Tokenize the text with words:
words = word_tokenize(text)
print(f"Number of words: {len(words)}")
print(words)
```

```python
from nltk.probability import FreqDist

# Find the frequency
fdist = FreqDist(words)

# Print 20 most common words
fdist.most_common(20)
```

```python
# Plot the grph for fdist
import matplotlib.pyplot as plt

fdist.plot(20)
```

```python
# Empty list to store words
words_no_punc = []

# Removing punctuation marks
for word in words:
    if word.isalpha():
        words_no_punc.append(word.lower())
        
# Print words without punctution marks and lenght
print(words_no_punc)
print(len(words_no_punc))
```

```python
# Plot graph without punctuation marks
fdist = FreqDist(words_no_punc)
print(fdist.most_common(20))

fdist.plot(20)
```

```python
"""Remove not useful words."""

# Download stopwords
from nltk.corpus import stopwords

print("Stopwprds:")
stopwords = stopwords.words("english")
print(stopwords)

# Remove stopwords from words_no_punc
clean_words = []

for word in words_no_punc:
    if word in stopwords:
        clean_words.append(word)
        
# Print clean words and lenght
print()
print(clean_words)
print(len(clean_words))
```

```python
"""Final frequency ditribution."""

# Plot graph clean words
fdist = FreqDist(clean_words)
print(fdist.most_common(20))

fdist.plot(20)
```

```python
"""WORLD CLOUD PYTHON IMPLEMENTATION."""

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generatingt the wordcloud
wordcloud = WordCloud().generate(text)

# Plot the wordcloud
plt.figure(figsize = (12,12))
plt.imshow(wordcloud)

#TO remove the axis value:
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud

# Use circle image as mask
char_mask = np.array(Image.open("circle.png"))

# Generatin wordcloud
wordcloud = WordCloud(background_color = "black", mask = char_mask).generate(text)

# Plot the wordcloud
plt.figure(figsize = (8,8))
plt.imshow(wordcloud)

# Remove the axis value
plt.axis("off")
plt.show()
```
