```python
# https://scikit-image.org/docs/stable/auto_examples/
# http://scipy-lectures.org/packages/scikit-image/

from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm

example_file = ('http://upload.wikimedia.org/' + 'wikipedia/commons/7/7d/Dog_face.png')
image = imread(example_file, as_gray = True)
plt.imshow(image, cmap = cm.gray)
plt.show()
```

```python
print(f"data type: {type(image)}, shape: {image.shape}")
```

```python
# First method to resize the image
image2 = image[5:70, 0:70]
plt.imshow(image2, cmap = cm.gray)
plt.show()
```

```python
# Second method to resizethe image - WRAP
image3 = resize(image, (30,30), mode="wrap")
plt.imshow(image3, cmap=cm.gray)
print(f"data type: {type(image3)}, shape: {image3.shape}")
```

```python
# Second method to resizethe image - EDGE
image3b = resize(image, (30,30), mode="edge")
plt.imshow(image3b, cmap=cm.gray)
```

```python
image_row = image3.flatten()
print(f"data type: {type(image_row)}, shape: {image_row.shape}")
```
